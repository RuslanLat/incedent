```python
# приведение в соответствие наименований
'ремонт подвальных помещений, относящихся к общему имуществу в многоквартирном доме' == 'ремонт подвальных помещений, относящихся к общему имуществу собственников помещений'
'ремонт подъездов, направленный на восстановление их надлежащего состояния и проводимый при выполнении иных работ по капитальному ремонту общего имущества в многоквартирном доме' ==
'ремонт подъездов, направленный на восстановление их надлежащего состояния и проводимый при выполнении иных работ'
'ремонт внутридомовых инженерных систем электроснабжения' == 'ремонт внутридомовых инженерных сетей электроснабжения'
'ремонт мусоропровода' == 'ремонт или замена мусоропровода'
'замена оконных блоков, расположенных в помещениях общего пользования' == 'ремонт фасада (замена оконных блоков, расположенных в помещениях общего пользования в многоквартирном доме)'
'ремонт внутреннего водостока' == 'ремонт или замена внутреннего водостока'
```

```python
def get_name(name):
    if name == 'ремонт подвальных помещений, относящихся к общему имуществу в многоквартирном доме':
        return 'ремонт подвальных помещений, относящихся к общему имуществу собственников помещений'
    elif name == 'ремонт подъездов, направленный на восстановление их надлежащего состояния и проводимый при выполнении иных работ по капитальному ремонту общего имущества в многоквартирном доме':
        return 'ремонт подъездов, направленный на восстановление их надлежащего состояния и проводимый при выполнении иных работ'
    elif name ==  'ремонт внутридомовых инженерных систем электроснабжения':
        return 'ремонт внутридомовых инженерных сетей электроснабжения'
    elif name == 'ремонт мусоропровода':
        return 'ремонт или замена мусоропровода'
    elif name == 'замена оконных блоков, расположенных в помещениях общего пользования':
        return 'ремонт фасада (замена оконных блоков, расположенных в помещениях общего пользования в многоквартирном доме)'
    elif name == 'ремонт внутреннего водостока':
        return 'ремонт или замена внутреннего водостока'
```

```python
# загрузка данных инциденты, зарегистрированные на объектах городского хозяйства
incidents_group_df = pd.read_csv("../data/incidents_group_df.csv")
incidents_group_df.head()
```

```python
# количество уникальных видов работ
overhaul_df["WORK_NAME"].nunique()
```

```python
# наименование работ
work_names = overhaul_df[["WORK_NAME"]].drop_duplicates().reset_index(drop=True).reset_index()
work_names
```

```python
# наименование видов обращений
incident_names = incidents_group_df[["Наименование"]].drop_duplicates().reset_index(drop=True)
incident_names
```

```python
# подготовка данных для merge
data = []
for i in range(len(incident_names)):
    for j in range(len(work_names)):
        data.append([j, incident_names.iloc[i].values[0]])
```

```python
# преобразование в DataFrame
incident_names = pd.DataFrame(data, columns=["index_work_names", "incident_names"])
incident_names.head()
```

```python
# размерность данных
incident_names.shape
```

```python
# объединение данных
incident_work_names = incident_names.merge(work_names, left_on="index_work_names", right_on="index").drop(columns=["index_work_names", "index"]).rename(columns={"WORK_NAME" : "work_names"})
incident_work_names.head()
```

```python
# размерность данных
incident_work_names.shape
```

```python
# сохранение данных в файл
incident_work_names.to_csv("../data/incident_work_names.csv", index=False)
```
