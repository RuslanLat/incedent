```python
# вывод таблицы 'systems'
db.Addresse.select().show()
```

```python
@db_session
def add_addresse(house_id, unom, adm_area, district, addresse_name):

    """ функуия записи адресов в таблицу - 'addresses' """
    
    db.Addresse(house_id=house_id, unom=unom, adm_area=adm_area,
                district=district, addresse_name=addresse_name)
    # commit() will be done automatically
    # database session cache will be cleared automatically
    # database connection will be returned to the pool
```

```python
# запись данных в базу данных
for unom, adm_area, district, addresse_name in addresse_desc:
    add_addresse(house_id=None, unom=unom, adm_area=adm_area,
                district=district, addresse_name=addresse_name)
```
