```python
def creat_predict_proba(pred_proba, rrr, key_dict):
    df_final = pd.DataFrame(data=pred_proba)
    dg = []
    for j in range(len(df_final)):
        for i in df_final.columns:
            data = {"umon" : rrr["Месяц создания во внешней системе"][j],
                    "Вероятность" : df_final[i][j].round(3),
                    "Наименование работ" : key_dict[str(i)]}
            dg.append(data)

    return dg
```

```python
dddd = creat_predict_proba(pred_proba, rrr, templates)
```
